Subplots
* Útil para criar gráficos multiplos em uma mesma figura

```python
np.random.seed(6)
x = np.arange(1,11)
y1 = np.random.randint(1,400,10)
y2 = np.random.randint(150,500,10)
y3 = np.random.randint(200,600,10)
```

```python
# cria figura
plt.figure(figsize=(15,5))
# cria o subplot 1
plt.subplot(1,3,1)
plt.plot(x, y1, color = 'k')
plt.title('Subplot 1')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
# cria o subplot 2
plt.subplot(1,3,2)
plt.plot(x, y2, color = 'r')

plt.title('Subplot 2')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')

# cria o subplot 3
plt.subplot(1,3,3)
plt.plot(x, y3, color = 'g')
plt.title('Subplot 3')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')

# ajusta espaçamento entre subplots
plt.tight_layout(pad=4)
plt.show()
```

```python
# outra forma de criar o subplot

# cria a figura e os eixos
from turtle import title


fig, ax = plt.subplots(1,3, figsize=(15,5))
fig.suptitle('Figure')

# line plot para cada eixo criado
ax[0].plot(x, y1, color='black')
ax[1].plot(x, y2, color='red')
ax[2].plot(x, y3, color='green')

# itera nos eixos e ajusta as labels
for i in range(3):
    ax[i].set(title = f'Subplot {i+1}', xlabel= 'Eixo x', ylabel = 'Eixo y')
```

```python
provas = ['math score', 'reading score','writing score']
fig, ax = plt.subplots(1,len(provas), figsize=(12,5))

for index, prova in enumerate(provas):
    mediana = students[prova].median()
    ax[index].hist(students[prova], edgecolor='black', alpha= 0.5)
    ax[index].axvline(mediana, color = 'black', ls= '--', label = 'Mediana')
    ax[index].legend()
    ax[index].set(xlabel = prova)
```
