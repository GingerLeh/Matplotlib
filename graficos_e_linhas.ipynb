Gráfico de linhas

Propriedades da linha e marcadores

* color ou c -> especificação de cor
* alpha -> espcificação de transparencia, numero float entre 0.0 e 1
* linewidth ou lw-> largura da linha
* linestyle ou ls -> estilo da linha -> -, --, :, -.
* marker -> especificação do marcador -> +, o, s
* markesize ou ms -> tamanho do marcador 
* marketedgecolor ou mec -> cor da linha do marcador 
* marketfacecolor ou mfc -> cor da face do marcador


```python
np.random.seed(7)
y = np. random.randint(low=1, high=1500, size=10)
```

```python
y
```

```python
plt.figure(figsize=(10,4))
plt.plot(y, color = 'black')
plt.title('Minha figura')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.show()
```

```python
plt.figure(figsize=(12,5))
# insere a primeira linha no plot
plt.plot(y, label='linha 1', color= 'black',
    marker = 'o', markersize = 2, markeredgecolor = 'red',
    linestyle='--'
)

# insere a segunda linha no plot
plt.plot(y*2, label= 'linha 2', color= 'red', 
    marker = 'D', ms = 4, mec='yellow', mfc = 'black', ls=':'
)

# rolulos
plt.xlabel('Eixo x', size=12)
plt.ylabel('Eixo y')
plt.title('Titulo', size= 20, loc='left', pad=10)

# gridlines
plt.grid(alpha= 0.2, color= 'darkgray', lw=1, axis= 'y')

# legenda
plt.legend()

plt.show()
```

```python
# exemplo format strings
plt.plot(y, 'o--k', color = 'r')
```

```python
mglu = yf.Ticker('MGLU3.SA').history(period='max')
```

```python
# ajusta o tamanho da figura
plt.figure(figsize=(12,4))

# line plot
plt.plot(mglu.index, mglu.Close, color='black',lw=1 )

# ajusta limites do eixo x
plt.xlim([mglu.index.min(),mglu.index.max()])
# ajusta limites do eixo y

# labels
plt.title('Ações MGLU', loc='left', pad=10, weight = 'bold')
plt.ylabel('Close price')
plt.xlabel('Tempo', size=10)

#grid lines
plt.grid(alpha = 0.2, ls='--' )

plt.show()
```
