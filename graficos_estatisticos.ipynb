```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
students = pd.read_csv('./materiais/StudentsPerformance.csv')
```

```python
# conta a frequencia de notas de forma ordenada
students['math score'].value_counts().sort_index(ascending=False).head()
```

```python
# sumario estatistico da nota de matematica
students['math score'].describe()
```

Gráficos estatísticos
* Histograma: evidencia a distribuiçã de frequencias de uma determinada variável 

```python
plt.hist(students['math score'])
```

```python
# calculo da mediana 
median_math = students['math score'].median()
plt.title('Histograma das notas em matemática')
plt.hist(students['math score'], edgecolor= 'black', alpha= 0.5)

# insere linha referente a nota mediana em matematica
plt.axvline(median_math, color = 'black', ls = '--', label='Mediana')

# ativa a legenda
plt.legend()

plt.show()
```

```python
# boxplot 
plt.boxplot(students['math score'])
plt.show()
```
