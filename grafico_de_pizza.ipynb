Gráfico de pizza
* Gráfico útil para demonstrar a contribuição de valores individuais relavivamente a um determinado valor total. Por exemplo, a contribuição de determinadas empresas para a receita total de um grupo empresarial. 

```python
nome_empresas = ['Empresa A','Empresa B','Empresa C']
receita = [100, 50, 300]
#cria o gráfico de pizza
plt.pie(receita, labels=nome_empresas)

plt.show()
```

```python
nome_empresas = ['Empresa A','Empresa B','Empresa C']
receita = [100, 50, 300]
colors = ['blue','red','green']
plt.pie(receita,
        labels = nome_empresas,
        colors = colors,
        wedgeprops={'linewidth':1, 'edgecolor':'white'} #propriedade gráficas
)

plt.show()
```
