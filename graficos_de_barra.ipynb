```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
mglu = yf.Ticker('MGLU3.SA').history(period='max')
```

Gráfico de barras

```python
# fazendo resample para pegar cotação média anual
mglu.resample('Y').mean().loc['2018':,'Close']
```

```python
# calculo da cotação média anual
mediaAnual = mglu.resample('Y').mean().loc['2018':,'Close']
```

```python
# cria o grafico de barras
plt.bar(mediaAnual.index.year,mediaAnual.values)

# limite do eixo y
plt.ylim(0,mediaAnual.values.max()+10)

# labels
plt.title('Cotação média anual', loc= 'left', pad=10, weight='bold')
plt.xlabel('Ano')
plt.ylabel('Preço de fechamento anual médio')
plt.show()
```
