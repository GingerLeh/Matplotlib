Scartterplot
* Gráfico de dispersão, para analisar a relação entre duas variáveis numéricas
* Podemos verificar se essa relação é linear ou não linear, neutra, positiva ou negativa
* É possivel adicionar uma terceira varicavel (categórica) para coloris os pontos ou ainda numerica (bubble plot ou gráfico de bolhas)

```python
# cria dados sintéticos
np.random.seed(7)
x = np.random.normal(10,0.5,100)
y = np.random.uniform(0,20,100)
```

```python
# cria scatter plot
plt.title('Grafico de dispersão')
plt.scatter(x,y, marker='o', color='red', label='data 1', alpha=0.5)
plt.scatter(x*0.5,y*0.5, color='black', marker='v', label='data 2', alpha= 0.5)
```
