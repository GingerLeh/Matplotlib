Bubble plot
* O grafico de bolhas acrescenta uma terceira dimensão ao grafico de dispersão

```python
# cria dados sintéticos
x = np.random.normal(loc=100, scale=0.5, size=50)
y = np.random.normal(loc=100, scale=2.5, size=50)
z = np.random.normal(loc=100, scale=3.5, size=50)
```

```python
# cria o bubble plot, adiciona uma terceira dimensão ao scatter plot
plt.scatter(x, y, alpha= 0.5, s = z*2)
plt.title('Bubble plot')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.show()
```
